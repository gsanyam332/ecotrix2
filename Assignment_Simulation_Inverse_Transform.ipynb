```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#Generating standard cauchy distribution
```

```python
#Generating uniform numbers
```

```python
uni = np.random.uniform(0, 1, 100000)
cauchy = []
for i in uni:
    sim_cauchy = np.tan(np.pi * (i - 0.5))
    cauchy.append(sim_cauchy)
    
cauch = np.array(cauchy)
print(cauch.shape)
```

```python
n, bins, patches = plt.hist(cauch, bins=100, 
                            density = 1,
                            color ='green')
   
y = 1/(np.pi*(1 + np.power(bins,2)))
  
plt.plot(bins, y, '--', color ='black')
  
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
  
plt.show()
```

```python
#Generating cauchy(a,b) with standard cauchy
```

```python
#input parameters here
a=2
b=3
uni = np.random.uniform(0, 1, 100000)
cauchy = []
for i in uni:
    sim_cauchy = 2 + 3*np.tan(np.pi * (i - 0.5))
    cauchy.append(sim_cauchy)
    
cauch = np.array(cauchy)
print(cauch.shape)

plt.figure(figsize=(10,10))
n, bins, patches = plt.hist(cauch, bins=200, 
                            density = 1,range=(a-30,a+30),
                            color ='green')
   
y = 1/(b*np.pi*(1 + np.power((bins-a)/b,2)))
  
plt.plot(bins, y, '--', color ='black')
  
plt.xlabel('X')
plt.ylabel('Density')
  
plt.show()
```
