```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#Generating truncated cauchy -500,500
```

```python
c = 1000/(np.pi*0.9987)
cauchy = []
count = 0
while(count<=2000):
    sim_uni = np.random.uniform(0, 1, 1)
    sim_uni = 1000*sim_uni - 500
    sim_uni2 = np.random.uniform(0, 1, 1)
    if(sim_uni2 <= 1/(1+np.power(sim_uni,2))):
        cauchy.append(sim_uni)
        count=count+1
    else:
        continue

cauch = np.array(cauchy)
plt.figure(figsize=(10,10))
n, bins, patches = plt.hist(cauch, bins=100, 
                            density = 1,range=(-20,20),
                            color ='green')
   
y = 1/(0.9987*np.pi*(1 + np.power(bins,2)))

  
plt.plot(bins, y, '--', color ='black')
  
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
  
plt.show()
```
